## Solution

As we find out from the Analysis that, Technology are doing maximum sales and also making maximum profit. So, in this, Category which have minimum sales and profit,we have to increase their supplies,then only we will get profit from them.
Secondly, The ship Mode which is Maximum use by customers are Same day and they are doing maximum sales but do not making any much profit. While First Class is least prefered and also doing less sales but they are making Maximum profit because the Shiping Price of First Calss is more than the other Ship Modes.Also, the supplies of technology is the main reason for growth of profit in First Class because the shiping of these product need much care as they are very costly , so,people generally prefered this type of ship mode.In order to get the more profit from the other ship modes, we have to increase the shiping Prices. 
Also, Homeoffice segment has done the maximum sales and making the maximum profit while consumers are the lest one.So, we have to decrease the suppies in Consumer Segment as they are not making much profit and increase the supplies in other two segments.
Also, the Region who have the more sales but do not making any profit.Either,we have to increase the selling price in those Regions or else reduce the supplies as they are not makeing much profit.
